```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('EDA_with_Avg.csv', index_col=0)
df.head()
```

```python
comorbidities_df = df[['outcome', 'age', 'gendera', 'hypertensive', 'atrialfibrillation', 'CHD with no MI', 
                       'diabetes', 'deficiencyanemias', 'depression', 'Hyperlipemia', 'Renal failure', 'COPD']]
comorbidities_df.head()
```

```python
comorbidities_df = comorbidities_df.rename(columns={'CHD with no MI':'CHD_with_no_MI', 'Renal failure': 'Renal_failure'})
comorbidities_df
```

```python
comorbidities_df.dtypes
```

```python
comorbidities_df.outcome = comorbidities_df.outcome.astype('bool')
comorbidities_df.hypertensive = comorbidities_df.hypertensive.astype('bool')
comorbidities_df.atrialfibrillation = comorbidities_df.atrialfibrillation.astype('bool')
comorbidities_df.CHD_with_no_MI = comorbidities_df.CHD_with_no_MI.astype('bool')
comorbidities_df.diabetes = comorbidities_df.diabetes.astype('bool')
comorbidities_df.deficiencyanemias = comorbidities_df.deficiencyanemias.astype('bool')
comorbidities_df.depression = comorbidities_df.depression.astype('bool')
comorbidities_df.Hyperlipemia = comorbidities_df.Hyperlipemia.astype('bool')
comorbidities_df.Renal_failure = comorbidities_df.Renal_failure.astype('bool')
comorbidities_df.COPD = comorbidities_df.COPD.astype('bool')
```

```python
comorbidities_df.dtypes
```

```python
comorb_list = ['hypertensive', 'atrialfibrillation', 'CHD_with_no_MI', 
                       'diabetes', 'deficiencyanemias', 'depression', 'Hyperlipemia', 'Renal_failure', 'COPD']
comorbidities_df['Total Comorbidities'] = comorbidities_df[comorb_list].sum(axis=1)
comorbidities_df
```

```python
comorbidities_df.to_csv('comorbidities.csv')
```
