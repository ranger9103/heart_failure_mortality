```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
```

```python
orig_df = pd.read_csv('data.csv')

# set the max columns to none
pd.set_option('display.max_columns', None)

orig_df.head()
```

```python
orig_df.describe()
```

```python
orig_df.info()
```

```python
tdf = orig_df[['PCO2', 'Bicarbonate']]
test_data = tdf[tdf['PCO2'].isnull()]
tdf.dropna(inplace=True)
y_train = tdf['PCO2']
X_train = tdf.drop('PCO2', axis =1)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit (X_train, y_train)
X_test = test_data.drop('PCO2', axis=1)
y_pred = lr.predict(X_test)
test_data.loc[test_data['PCO2'].isnull(), 'PCO2'] = y_pred
df_final = tdf.combine_first(test_data)
```

```python
orig_df.drop(['PCO2'], axis=1)
orig_df['PCO2'] = df_final['PCO2']
```

```python
orig_df.info()
```

```python
# one 'Outcome' row has a NaN value, removed due to not knowing if patient died or not
drop_df = orig_df.dropna(subset=['outcome'])
```

```python
drop_df.info()
```

```python
df = drop_df.fillna(drop_df.mean())
df.head()
```
